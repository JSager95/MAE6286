```python
import numpy
```

```python
from matplotlib import pyplot
```

```python
%matplotlib inline
```

```python
#Create a time grid
T = 100.0
dt = 0.02
N = int(T/dt)
t = numpy.linspace(0.0,T,num=N)
```

```python
#Set up intital conditions
z0 = 100.0 #Height
b0 = 10.0  #Upward velocity
zt = 100.0  #Trim height
g = 9.81    #Gravity

#Initial value of the numerical solution
u = numpy.array([z0,b0])

#Need to create an empty array to store all of the computed elevation levels
z = numpy.zeros(N)
z[0] = z0
```

```python
for n in range(1,N):
    rhs = numpy.array([u[1],g * (1- u[0] / zt)])
    u = u + dt * rhs
    z[n] = u[0]
    
```

```python
pyplot.title("Elevataion of the phugoid over time")
pyplot.xlabel("Time")
pyplot.ylabel("Height")
pyplot.xlim(t[0],t[-1])
pyplot.ylim(40.0,160.0)
pyplot.grid()
pyplot.plot(t,z, linestyle = "-", linewidth =2);
```
